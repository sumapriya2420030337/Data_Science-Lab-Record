```python
import numpy as np
from scipy import stats
b=np.array([120,122,118,130,125,128,115,121,123,119])
a=np.array([115,120,112,128,122,125,110,117,119,114])
alpha=0.05
t_stats,p_val=stats.ttest_rel(a,b)
m=np.mean(a-b)
s=np.std(a-b,ddof=1)
n=len(b)
t_manual=m/(s/np.sqrt(n))
decision="Rejects" if p_val<=alpha else "Fail to accept"
concl="Significant difference"if decision=="Rejects" else "No significant difference"
print("T",t_stats)
print("P",p_val)
print("T manual",t_manual)
print(f"Decision {decision} H0 at {alpha}")
print("Conclusion",concl)
```

```python
import numpy as np
from statsmodels.stats.weightstats import ztest
data=np.random.normal(loc=11.8,scale=0.5,size=100)
population_mean=12
population_std_dev=0.5
z_statistics,p_value=ztest(data,value=population_mean)
print(f"Z statistics:{z_statistics:.4f}")
print(f"P value:{p_value:.4f}")
alpha=0.05
if p_value<alpha:
  print("Rejects")
else:
  print("Accepted")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

data=[[20,25],
      [15,20],
      [10,30]]
chi2,p,dof,expected=chi2_contingency(data)
print(f"Chi Square Value {chi2:.3f}")
print(f"P-Value {p}")
print(f"Degrees of Freedom {dof}")
```
