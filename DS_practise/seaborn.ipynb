```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
fmri = sns.load_dataset("fmri")

sns.lineplot(data=fmri, x="timepoint", y="signal")
plt.show()
```

```python
sns.barplot(data=tips, x="day", y="total_bill")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", bins=20)
plt.show()
```

```python
sns.violinplot(data=tips, x="day", y="total_bill")
plt.show()
```

```python
sns.countplot(data=tips, x="day")
plt.show()
```

```python
import numpy as np

data = np.random.rand(5, 5)
sns.heatmap(data, annot=True)
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
corr = tips.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.kdeplot(data=tips, x="total_bill", fill=True)
plt.show()
```

```python
sns.regplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
sns.set_style("darkgrid")

sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
df=sns.load_dataset("tips")
sns.pairplot(df)
plt.show()
```

```python
df=sns.load_dataset("tips")
sns.pairplot(df,hue="sex")
plt.show()
```

```python
df=sns.load_dataset("iris")
cuspal={"setosa":"red","versicolor":"green","virginica":"blue"}
sns.scatterplot(data=df,x="sepal_length",y="sepal_width",hue="species",palette=cuspal)
plt.show()

```

```python
data=[12,15,20,20,22,23,25,25,25,30,32,35,40]
plt.hist(data,bins=5,color='skyblue',edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.show()

```

```python
df=sns.load_dataset('iris')
sns.violinplot(x='species',y='petal_width',data=df)
plt.title('violin plot')
plt.show()
```
