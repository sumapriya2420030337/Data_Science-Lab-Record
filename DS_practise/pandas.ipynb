```python
import pandas as pd

df = pd.DataFrame({
    'Date': ['2025-01-05', '05/01/2025', 'Jan 5, 2025', '2025.01.05']
})

df['Date'] = pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%Y-%m-%d')

print(df)
```

```python
df = pd.DataFrame({
    'Name': ['Alice', 'BOB', 'charlie', 'DAVID']
})

# Convert all names to lowercase
df['Name_lower'] = df['Name'].str.lower()
print(df)

# Convert all names to uppercase
df['Name_upper'] = df['Name'].str.upper()

print(df)
```

```python
iris_df = pd.read_csv('Iris.csv')

print("Original Iris DataFrame:")
display(iris_df.head())

iris_df['Species_lower'] = iris_df['Species'].str.lower()

iris_df['Species_upper'] = iris_df['Species'].str.upper()

print("\nIris DataFrame with lowercase and uppercase Species:")
display(iris_df[['Species', 'Species_lower', 'Species_upper']].head())
```
